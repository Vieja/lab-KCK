```python
coins = data.coins()
coins
def thresh(t):
    #warnings.simplefilter("ignore")
    binary = (img > t) * 255
    #binary = np.uint8(binary)
    io.imshow(binary)
    flush_figures()

interact(thresh, t=(0,255,1))
```

```python
img = exposure.adjust_gamma(img, 2)
io.imshow(img)
```

```python
MIN = np.percentile(img, 0)
MAX = np.percentile(img, 10)
img = (img - MIN) / (MAX - MIN)
img[img > 1] = 1
img[img < 0] = 0
io.imshow(img)
```
