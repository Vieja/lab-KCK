```python
import matplotlib
import colorsys
from matplotlib import pyplot as plt
from matplotlib import colors
import numpy as np
import math

def main():
    wysokosci = []
    pomap = []
    map = []
    angles = []
    min_val = 0.432528741577922
    max_val = 1.5383204012924698
    min_angle = 0
    max_angle = 0
    wektor_swiatla = (0,1000,1000)
    
    with open('big.dem') as f:
        w, h, d = f.readline().split()

        for line in f:
            wys = []
            pom = []
            for item in line.split():
                pom.append((float(item) / 100 - min_val) / (max_val - min_val))
                wys.append(float(item))
            pomap.append(pom)
            wysokosci.append(wys)

        for linijka in pomap:
            pom = []
            for wysokosc in linijka:
                pom.append(get_gradient(wysokosc))
            map.append(pom)
        
        for i in range(1,int(h)):
            for j in range(0,int(w)-1):
                p1 = np.array([0, 0, wysokosci[i][j]])
                p2 = np.array([0, 1, wysokosci[i-1][j]])
                p3 = np.array([1, 0, wysokosci[i][j+1]])
                v1 = p3 - p1
                v2 = p2 - p1
                cp = np.cross(v1, v2)
                skalarny = np.dot(cp,wektor_swiatla)
                iloczyn = np.linalg.norm(cp) * np.linalg.norm(wektor_swiatla)
                cosinus = skalarny / iloczyn
                #print("cosinus",cosinus)
                if (cosinus < 0):
                    cosinus *= -1
                    rad=math.acos(cosinus)
                    angle=math.degrees(rad)
                    angle *= -1
                    #angles.append(angle) 
                else:
                    rad=math.acos(cosinus)
                    angle=math.degrees(rad)
                r, g, b = adjust_light(map[i][j][0],map[i][j][1],map[i][j][2],angle)
                map[i][j] = (r, g, b)
    
    plt.figure(figsize=(5,5))
    plt.imshow(map)
    plt.savefig('my-map.pdf')

def adjust_light(r, g, b, angle):
    h, s, v = rgb2hsv(r, g, b)
    #print(s)
    if (angle < 45):
        s = 1 - angle/45
        #print(s)
    r, g, b = hsv2rgb(h, s, v)
    return r, g, b

def rgb2hsv(r, g, b):
    h, s, v = colorsys.rgb_to_hsv(r, g, b)
    return (h, s, v)

def hsv2rgb(h, s, v):
    r, g, b = colorsys.hsv_to_rgb(h, s, v)
    return (r, g, b)

def get_gradient(i):
    if (i < 9/32):   # ziel
        r = i * 2.5
        g = 1
        b = 0
    elif (i < 1 / 2): # zolty
        r = 0.2 + i * 1.5
        g = 1
        b = 0 
    else:           # czerwony
        r = 1
        g = 1 - (i - 1/2) * 2 
        b = 0
    return (r, g, b)

if (__name__ == '__main__'):
    main()
```
