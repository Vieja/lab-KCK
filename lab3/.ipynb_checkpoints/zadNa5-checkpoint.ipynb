```python
import matplotlib
import colorsys
from matplotlib import pyplot as plt
from matplotlib import colors
import numpy as np
import math

def main():
    wysokosci = []
    pomap = []
    map = []
    min_val = 0.432528741577922
    max_val = 1.5383204012924698
    wektor_swiatla = (0,1000,1000)
    
    with open('big.dem') as f:
        w, h, d = f.readline().split()

        for line in f:
            wys = []
            pom = []
            for item in line.split():
                pom.append((float(item) / 100 - min_val) / (max_val - min_val))
                wys.append(float(item))
            pomap.append(pom)
            wysokosci.append(wys)

    for linijka in pomap:
        pom = []
        for wysokosc in linijka:
            pom.append(get_gradient(wysokosc))
        map.append(pom)

    for i in range(1,int(h)-1):
        for j in range(1,int(w)-1):
            v1 = np.array([10, 0, wysokosci[i][j+1]-wysokosci[i][j] ])
            v2 = np.array([0, 10, wysokosci[i-1][j]-wysokosci[i][j] ])
            v3 = np.array([-10, 0, wysokosci[i][j-1]-wysokosci[i][j] ])
            v4 = np.array([0, -10, wysokosci[i+1][j]-wysokosci[i][j] ])
            suma_wektorow = np.cross(v1,v2)+np.cross(v2,v3)+np.cross(v3,v4)+np.cross(v4,v1)
            wektor_norm = normalize(suma_wektorow)
            
            skalarny = np.dot(wektor_norm,wektor_swiatla)
            iloczyn = np.linalg.norm(wektor_norm) * np.linalg.norm(wektor_swiatla)
            cosinus = skalarny / iloczyn
            rad=math.acos(cosinus)
            angle=math.degrees(rad)
            if (angle > 45):
                r, g, b = light_down(map[i][j][0],map[i][j][1],map[i][j][2],angle)
                map[i][j] = (r, g, b)
            else:
                r, g, b = light_up(map[i][j][0],map[i][j][1],map[i][j][2],angle)
                map[i][j] = (r, g, b)
                
    plt.figure(figsize=(5,5))
    plt.imshow(map)
    plt.savefig('mapka3.jpg', quality=100)
    
def crossvectors(a,b,c,d):
    return normalize(np.cross(a,b)+np.cross(b,c)+np.cross(c,d)+np.cross(d,a))

def normalize(x):
    x = x/(math.sqrt((x[0] **2) + (x[1] **2) + (x[2] **2)))
    return x

def light_down(r, g, b, angle):
    h, s, v = rgb2hsv(r, g, b)
    angle -= 40
    v = min(1.1 - max(angle/60.2, 0.2) , 1)
    r, g, b = hsv2rgb(h, s, v)
    return r, g, b

def light_up(r, g, b, angle):
    h, s, v = rgb2hsv(r, g, b)
    angle = math.fabs(angle-22.5)
    s = pow(angle/22.5,1/2)
    r, g, b = hsv2rgb(h, s, v)
    return r, g, b

def rgb2hsv(r, g, b):
    h, s, v = colorsys.rgb_to_hsv(r, g, b)
    return (h, s, v)

def hsv2rgb(h, s, v):
    r, g, b = colorsys.hsv_to_rgb(h, s, v)
    return (r, g, b)

def get_gradient(i):
    if (i < 9/32):   # zielony
        r = i * 2.5
        g = 1
        b = 0
    elif (i < 1 / 2): # zolty
        r = 0.2 + i * 1.5
        g = 1
        b = 0 
    else:           # czerwony
        r = 1
        g = 1 - (i - 1/2) * 2 
        b = 0
    return (r, g, b)

if (__name__ == '__main__'):
    main()
```
