```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from sympy import *

import sympy as sym
import numpy as np

%matplotlib inline
from matplotlib import pyplot as plt
```

# Sympy

```python
x, y = sym.symbols("x y")

eq = ((-x)**3 + 3*x**2 + 10*x - 24)
eq
```

```python
lista = np.arange(-5,5,0.1)
wartosci = []
for z in lista:
    wartosci.append(eq.evalf(subs = {x : z}))
plt.plot(lista, wartosci)
plt.show()
```

```python
solve(eq,x)
```

```python
eq1 = (x**2 + 3*y - 10)
eq2 = (4*x - y**2 + 2)
wynik = solve([eq1,eq2])
wynik
```

```python
print("Liczba rozwiązań = ", len(wynik))
```

```python
print("Rozwiązanie w postaci numerycznej:")
for s in wynik:
     for x, y in s.items():
        print(x, y.evalf())
```
