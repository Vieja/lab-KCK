```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['font.size'] = 13

csv=('rsel.csv','cel-rs.csv','2cel-rs.csv','cel.csv','2cel.csv')
nazwy=('1-Evol-RS','1-Coev-RS','2-Coev-RS','1-Coev','2-Coev')
symbole=('o','v','D','s','d')
kolory=('blue','green','red','black','violet')

def wczytajPlik(plik):
    arr = []
    with open(plik) as f:
        arr = [line.split(',') for line in f.readlines()[1:]]
    return arr

def ostatniePokolenie(a):
    arr = []
    for i in a[-1][2:]:
        arr.append(float(i)*100)
    return arr

plots = plt.figure(figsize=(10,7))
first = plots.add_subplot(121)

tabele = []
for dane, nazwa, sym, col in zip(csv,nazwy,symbole,kolory):
    cel = wczytajPlik(dane)
    tabele.append(cel)
    cel_avg = []
    cel_pop = []
    for gen in cel:
        suma = 0
        for item in gen[2:]:
            suma+=float(item)
        cel_avg.append((suma/(len(gen)-2))*100)
        cel_pop.append(int(gen[1])/1000)
    first.plot(cel_pop,cel_avg,label=nazwa,marker=sym,markersize=7,markevery=25,c=col,markeredgecolor='black')

first.set_xlabel("Rozegranych gier (x1000)")
first.set_ylabel("Odsetek wygranych gier [%]")
first.legend(loc='lower right',numpoints=2,markerscale=0.9)
first.grid(axis='both',linestyle=':')
first.set_xlim(0,500)
first.set_ylim(60,100)

top = plt.twiny()
top.set_xlabel("Pokolenie")
top.set_xticklabels(["0","40","80","120","160","200"])

second = plots.add_subplot(122)
boxdata = []
for dane in tabele:
    boxdata.append(ostatniePokolenie(dane))
    
box = second.boxplot(boxdata,1,'b+', showmeans=True)
plt.setp(box['boxes'], color='blue')
plt.setp(box['whiskers'], color='blue', linestyle='-.')
plt.setp(box['means'], markerfacecolor='blue', marker='o', markeredgecolor='black')
second.grid(axis='both',linestyle=':')
second.set_ylim(60,100)
second.set_xticklabels(nazwy, rotation=20)
second.yaxis.tick_right()
    
plt.savefig('wizualizacja.pdf')
plt.close()
```
